```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Pregunta 1

df_marcas = pd.read_csv('Marcas.csv', encoding = 'windows-1250')
df_modelos = pd.read_csv('Modelos.csv', encoding = 'windows-1250')
```

```python
#Pregunta 2

df = pd.merge(left= df_marcas, right= df_modelos, on= 'ID_Marca')
```

```python
#Pregunta 3

df.tail(5)
```

```python
#Pregunta 4

#Utiliza los filtros para mostrar todos los teléfonos con 64GB en la columna ROM_GB e indica cuántos del listado poseen esta característica.
df[(df['ROM_GB'] == 64)]
```

```python
#Pregunta 5

#Haz un histograma (hist) de la columna ROM_GB del data frame df.
#El gráfico debe incluir un título y las columnas tener un color personalizado.
fig, ax = plt.subplots(figsize= (10,5))
x = df['ROM_GB']
ax.hist(x, color= 'orange', bins=10, edgecolor= 'white')
ax.set_title('Resultados de celulares')
ax.set_xlabel('XLabel')
ax.set_ylabel('YLabel')
plt.show()
```

```python
#Pregunta 6

#Haz un diagrama de dispersión (scatter) con las siguientes características:
#el eje X deberá indicar el contenido de la columna Precio de lista, mientras que el eje
#el eje Y expresará el contenido de la columna Precio de venta.
#El diagrama debe incluir un título y los puntos tener un color personalizado.

fig, ax = plt.subplots()
x = df['Precio de lista']
y = df['Precio de venta']
ax.scatter(x, y, color='green')
ax.set_title('Precio de lista y precio de venta de celulares con 64 GB')
ax.set_xlabel('Precio de lista')
ax.set_ylabel('Precio de venta')
plt.show()
```
